```python
erro = 10**(-4)
f = [1, 0, -9, 3]
df = [3, 0,-9]
x0 = 1.5
x = erro
```

```python
def newton(f, df, x0):
  num = []
  den = []
  numSum = 0
  denSum = 0
  indice = 0
  for i in range(len(f)-1):
    num += [((f[i]*x0)**(len(f)-indice - 1))]
    indice = indice+1
  num += [f[len(f)-1]]
  indice = 0
  for j in range(len(df)-1):
    den += [((df[j]*x0)**(len(df)-indice - 1))]
    indice = indice+1
  den += [df[len(df)-1]]
  for i in range(len(num)):
    numSum += num[i]
  for j in range(len(den)):
    denSum += den[j]
  x = x0 - (numSum/denSum)


  print("erro: ", erro)
  print("x0: ", x0)
  print("valor de x: ", x)
  if abs(x0 - x) > erro:
    print("erro calculado: ", abs(x0 - x))
    print("------------------------")
    return (newton(f,df,x))



```
