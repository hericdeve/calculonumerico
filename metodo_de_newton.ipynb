```python
erro = 10**(-4)
x0 = 1.5
```

```python
def g(x):
    return x**3-9*x+3
```

```python
def dg(x):
    return 3*x**2-9
```

```python
def newton(f, df, x0):
  x = x0 - (f(x0)/df(x0))
  print("erro: ", erro)
  print("x0: ", x0)
  print("valor de x: ", x)
  if abs(x0 - x) > erro:
    print("erro calculado: ", abs(x0 - x))
    print("------------------------")
    return (newton(f,df,x))
  
```

```python
newton(g,dg,x0)
```
