```python
sis = [[10, 2, 1, 7], [1, 5, 1, -8], [2, 3, 10, 6]]
x0 = [[7/10],[-8/5],[6/10]]
erro = 0.05
```

```python
def addMatriz(A, B):
    result = []
    for i in range(len(A)):
        result += [[]]
        for j in range(len(A[0])):
            result[i] += [A[i][j] + B[i][j]]    
    return result    
```

```python
def multiMatriz(A,B):

  # A = [[0, -0.2, -0.1], [-0.2, 0, -0.2], [-0.2, -0.3, 0]]
  # B = [[0.7], [-1.6], [0.6]]

  tamanhoA = len(A)
  tamanhoB = len(B[0])
  result = []

  for i in range(tamanhoA):
    result += [[]]
    for j in range(tamanhoB):
      result[i] += [0]


  for i in range(len(A)):
    for j in range(len(B[0])):
        for k in range(len(B)):
            result[i][j] += A[i][k] * B[k][j]
  return (result)
```

```python
def calculaErro(x,x0):
  result = []
  for i in range(len(x)):
      result += [[]]
      for j in range(len(x[0])):
          result[i] += [x[i][j] - x0[i][j]]
  return result
```

```python
def isola(x, coe):
  b = x[-1]
  xi = []

  for i in range(len(x)-1):
    if i != coe:
      xi += [-1*(x[i]/x[coe])]
    elif i == coe:
      xi += [0]
  xi += [b/x[coe]]
  return(xi)
```

```python
def gauss_jacobi(sis,x0,erro):
  x = []
  for i in range(len(sis)):
    x += [isola(sis[i],i)]
  
  b = []
  for i in range(len(x)):
    b += [[x[i][-1]]]
  x = [linha[:-1] for linha in x]
  print(x)
  print(b)
  x1 = addMatriz(b,multiMatriz(x,x0))
  print(x1)
  print("erro", calculaErro(x1,x0))
  print("max erro = ",max(abs(erro[0]) for erro in (calculaErro(x1,x0))))
  print("----------------------")
  if(max(abs(erro[0]) for erro in (calculaErro(x1,x0))) > erro):
    return gauss_jacobi(sis,x1,erro)
  return x1
  
  
```
