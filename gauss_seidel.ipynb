```python
def addMatriz(A, B):
    result = []
    for i in range(len(A)):
        result += [[]]
        for j in range(len(A[0])):
            result[i] += [A[i][j] + B[i][j]]    
    return result    
```

```python
def iteraMatriz(x,x0_dentro, b):

  x0 = copy.deepcopy(x0_dentro)
  print("x0: ", x0)
  
  tamanhoA = len(x)
  tamanhoB = len(x0[0])
  result = []

  for i in range(tamanhoA):
    result += [[]]
    for j in range(tamanhoB):
      result[i] += [0]


  for i in range(len(x)):
      for j in range(len(x0[0])):
    
          for k in range(len(x0)):
              result[i][j] += x[i][k] * x0[k][j]
      
      # print("result antes soma: ", result)
      # print("result of x{} = {} + {}".format(i+1, b[i][0], result[i][0]))
    
      result[i][0] = b[i][0] + result[i][0]

      x0[i][0] = result[i][0]
      # print("result: ", result)
      # print("x0: ", x0)
      # print("-------------")
  print("x0_final: ", x0)
  return (result)
```

```python
def calculaErro(x,x0):
  result = []
  for i in range(len(x)):
      result += [[]]
      for j in range(len(x[0])):
          result[i] += [x[i][j] - x0[i][j]]
  return result
```

```python
def isola(x, x0, coe):
  b = x[-1]
  xi = []

  for i in range(len(x)-1):
    if i != coe:
      xi += [-1*(x[i]/x[coe])]
    elif i == coe:
      xi += [0]
    # x0[i] = xi[-1]
  xi += [b/x[coe]]
  return(xi)
```

```python
def gauss_seidel(sis,x0,erro):
  print("x0 dentro de gauss: ", x0)
  x = []
  for i in range(len(sis)):
    x += [isola(sis[i],x0,i)]
  
  b = []
  for i in range(len(x)):
    b += [[x[i][-1]]]
  x = [linha[:-1] for linha in x]
  print("x: ", x)
  print("b: ", b)

  print("(x,x0,b): ", iteraMatriz(x,x0, b))
  x1 = iteraMatriz(x,x0, b)
  print("x1: ",x1)
  print("x0 dentro de gauss depois de itera: ", x0)
  print("erro", calculaErro(x1,x0))
  print("max erro = ",max(abs(erro[0]) for erro in (calculaErro(x1,x0))))
  print("----------------------")
  if(max(abs(erro[0]) for erro in (calculaErro(x1,x0))) > erro):
    return gauss_seidel(sis,x1,erro)
  return x1
```
